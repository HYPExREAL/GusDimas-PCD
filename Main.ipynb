```python
from ultralytics import YOLO
import cv2
import os

model = YOLO("runs/detect/train6/weights/best.pt")

#img_path = "archive/dataset/train/images/id_2_2_train.jpg"
img_path = "3.jpg"
results = model(img_path, conf=0.8)

os.makedirs("runs/detect/output", exist_ok=True)

for result in results:
    img = result.orig_img.copy()
    boxes = result.boxes.xyxy
    labels = result.boxes.cls
    confs = result.boxes.conf


    for box, label, conf in zip(boxes, labels, confs):
        x1, y1, x2, y2 = map(int, box)
        cls_name = model.names[int(label)]
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(img, f"{cls_name} {conf:.2f}", (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 0), 2)

    save_path = "runs/detect/output/detected.jpg"
    cv2.imwrite(save_path, img)
    print(f"Hasil disimpan ke: {save_path}")

    cv2.imshow("Deteksi Helm", img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
